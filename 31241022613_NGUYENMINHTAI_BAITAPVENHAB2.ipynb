```python
def analyze_market_strategy(market_demand, rival_impact, inventory):
    strategy = "Duy trì mức giá hiện tại"

    # Trường hợp nhu cầu cao và ít đối thủ
    if market_demand > 80 and rival_impact < 30:
        strategy = "Điều chỉnh tăng giá hoặc thêm quà tặng"

    # Trường hợp tồn kho lớn nhưng nhu cầu thấp
    elif inventory > 100 and market_demand < 40:
        strategy = "Khuyến mãi xả kho (Giảm 40%)"

    # Trường hợp áp lực cạnh tranh quá lớn
    elif rival_impact > 70:
        strategy = "Chạy chương trình Flash Sale"

    return f"Kế hoạch hành động: {strategy}"

# Chạy thử với bộ số liệu mới: Nhu cầu 50, Áp lực cạnh tranh 85, Tồn kho 120
print(analyze_market_strategy(50, 85, 120))
```

```python
def tinh_chiet_khau_san(sao_danh_gia, luong_ban, loi_nhuan, co_su_kien):
    # Khởi tạo giá trị giảm giá ban đầu
    tile_giam_gia = 0

    # Nhóm điều kiện cho shop uy tín, doanh số tốt
    if sao_danh_gia > 4.5 and luong_ban > 500 and loi_nhuan > 30:
        tile_giam_gia = 5

    # Nhóm điều kiện cho shop cần đẩy đơn (rating thấp, ít khách)
    elif sao_danh_gia < 4.0 and luong_ban < 100:
        tile_giam_gia = 20

    # Cộng thêm ưu đãi nếu đang trong đợt Sale (11.11, 12.12...)
    if co_su_kien:
        tile_giam_gia += 10

    return f"Tỷ lệ giảm giá cuối cùng: {tile_giam_gia}%"

# Kiểm tra với bộ thông số mới: Rating 4.8, Bán được 600 đơn, Lời 35%, Có sự kiện
print(tinh_chiet_khau_san(4.8, 600, 35, True))
```

```python
def grab_bike_calculator(dist, traffic, req_count, weather_status):
    # Khởi tạo giá trị mặc định
    res = {"p": "Trung bình", "b": "Trung bình"}

    if dist < 3 and traffic < 30 and req_count < 30:
        res["p"] = "Thấp"
    elif dist > 10 and traffic > 70:
        res["p"] = "Rất cao"

    if weather_status == "Mưa to" or req_count > 80:
        res["p"], res["b"] = "Cao", "Cao"

    return f"Giá cước: {res['p']}, Điểm thưởng: {res['b']}"

print(grab_bike_calculator(3, 40, 75, "Nắng"))
```

```python
def phan_bo_don_hang(mat_do_don, tai_xe_dang_gan, tinh_trang_gt):
    # Thiết lập số lượng đơn mặc định
    so_don_assign = 1

    # Ưu tiên giao nhiều đơn khi khu vực tập trung và đường thoáng
    if mat_do_don > 80 and tai_xe_dang_gan < 2 and tinh_trang_gt < 40:
        so_don_assign = 5
    # Hạn chế đơn khi giao thông quá tệ (tắc đường)
    elif tinh_trang_gt > 80:
        so_don_assign = 1
    # Các trường hợp bình thường khác
    else:
        so_don_assign = 3

    return f"Số lượng đơn cần giao: {so_don_assign}"

# Chạy thử với bộ số mới: Mật độ 70, Tải xế đang nhận 1, Giao thông 50
print(phan_bo_don_hang(70, 1, 50))
```
